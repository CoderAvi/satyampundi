```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x=np.arange(0,3*np.pi,0.1)
y=np.sin(x)
plt.plot(y)
plt.show()
```

```python
#sin wave
x=np.arange(0,3*np.pi,0.1)
ysin=np.sin(x)
ycos=np.cos(x)
plt.subplot(2,1,1)
plt.plot(x,ysin)
plt.title('sin')
plt.subplot(212)
plt.plot(x,ycos)
plt.title('cos')
plt.show()
```

```python
x=np.arange(0,3*np.pi,0.1)
y=np.sin(x)
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_title('sine')
ax.set_xlabel('angle')
ax.set_ylabel('sine')
plt.show()
```

```python
x=np.arange(0,3*np.pi,0.1)
y=np.sin(x)
fig=plt.figure()
ax1=fig.add_axes([0,0,1,1])#main axes
ax2=fig.add_axes([0.55,0.55,0.3,0.3])#insert axes
ax1.plot(x,y,'b')
ax2.plot(x,np.cos(x),'r')
ax1.set_title('sine')
ax2.set_title('cos')
plt.show()
```

```python
fig=plt.figure()
ax2=fig.add_subplot(111)
ax2.plot([1,2,3])
ax1=fig.add_subplot(221,facecolor='y')
ax1.plot([1,2,3])

```

```python
#subplots figure object and array of axes objects
x=np.arange(1,10,1)
fig,a=plt.subplots(2,2)
a[0][0].plot(x,x*x)
a[0][0].set_title('sq')
a[0][1].plot(x,np.exp(x))
a[0][1].set_title('ep')
a[1][0].plot(x,np.log(x))
a[1][0].set_title('log')
a[1][1].plot(x,np.sqrt(x))
a[1][1].set_title('sqrt')
plt.show()
```

```python
fig,ax=plt.subplots(1,3,figsize=(12,4))
ax[0].plot(x,x*x)
ax[1].plot(x,np.exp(x))
ax[2].plot(x,np.sqrt(x))
fig.tight_layout()
plt.show()
```

```python
a1=plt.subplot2grid((3,3),(0,0),colspan=2)
a2=plt.subplot2grid((3,3),(0,2),rowspan=3)
a3=plt.subplot2grid((3,3),(1,0),rowspan=2,colspan=2)
a1.plot(x,x*x)
a2.plot(x,np.exp(x))
a3.plot(x,np.log(x))
plt.tight_layout()#no overlapping issue
plt.show()
```

```python
fig=plt.figure()
a1=fig.add_axes([0,0,1,1])
a1.plot(x,np.exp(x))
a1.set_xlim(0,10)
a1.set_ylim(0,1000)
plt.show()
```
