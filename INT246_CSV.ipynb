```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
bmark=[10,20,20,11,23]
gmark=[13,24,23,11,43]
ind =[1,2,3,4,5]
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
width=0.45
ax.bar(ind,bmark,width,color='r')
ax.bar(ind,gmark,width,color='b',bottom=bmark)

ax.set_ylabel('marks')
ax.set_xticks(ind)
ax.set_xticklabels(['g1','g2','g3','g4','g5'])
ax.set_yticks(np.arange(0,100,10))
ax.legend(loc=2,labels=['boys','girls'])
plt.show()
```

```python
a=[12,21,12,1,13]
s=['a','b','c','d','e']
plt.pie(a,labels=s,autopct="%.f%%")
plt.show()
```
